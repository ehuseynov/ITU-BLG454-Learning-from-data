## Here is some terminology reminder for me
### Note: all the information below is from the internet
- **Epoch**: In the context of training a machine learning model, an epoch refers to a single pass through the entire training dataset. For example, if the training dataset contains 1000 examples, and the model is trained using 10 epochs, this means that the model will see each example 10 times, for a total of 10000 examples. **(another word for iteration )**
- **Propagation**: It refers to the process of computing the output of a neural network given an input. There are two types of propagation: forward propagation and backward propagation.

    - Forward propagation is the process of computing the output of a neural network given an input, using the current values of the weights and biases. <br>
    _In my code, forward propagation is used to compute the output of the MLP (A2) given the input data (X_training)._
    - Backward propagation is the process of computing the gradients of the loss with respect to the weights and biases of a neural network. <br>
    _In my code, backward propagation is used to compute the gradients of the loss with respect to the weights and biases of the MLP (dW1, db1, dW2, and db2)._
- **Activation**: It refers to a function that is applied to the output of a layer in a neural network. Activation functions are used to introduce non-linearity into the model, which allows the model to learn more complex patterns in the data. <br>
_In my implementation, the sigmoid activation function is used in the hidden layer, and the identity activation function is used in the output layer._



```python
# IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# LOAD NECESSARY DATA
training_df = pd.read_csv('data/optdigits.tra', header=None)
X_training, y_training = training_df.loc[:, 0:63], training_df.loc[:, 64]
print("Training Data:")
print("x training:")
print(X_training.shape)
print(X_training)
print("y training:")
print(y_training)

testing_df = pd.read_csv('data/optdigits.tes', header=None)
X_testing, y_testing = testing_df.loc[:, 0:63], testing_df.loc[:, 64]
print("Testing Data:")
print("x testing:")
print(X_testing.shape)
print(X_testing)
print("y testing:")
print(y_testing)
```

```python
# SCALE THE DATA
X_training = X_training / 16.0
X_testing = X_testing / 16.0

# the architecture of the autoencoder
hidden1_size = 32
hidden2_size = 16
input_size = 64
output_size = input_size

# initialization of the weights and biases (weight: random between 0 and 1, bias = 0)
W1 = np.random.normal(0, 1, size=(input_size, hidden1_size))
b1 = np.zeros(hidden1_size)
W2 = np.random.normal(0, 1, size=(hidden1_size, hidden2_size))
b2 = np.zeros(hidden2_size)
W3 = np.random.normal(0, 1, size=(hidden2_size, output_size))
b3 = np.zeros(output_size)
```

```python
# DEFINITON OF NECESSARY FUNCTIONS
# sigmoid function - sigmoid(x) = 1 / (1 + e^(-x))
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# derivative of the sigmoid function - sigmoid'(x) = sigmoid(x) * (1 - sigmoid(x)) 
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))


# linear equation function - linear(x) = x * w + b
def linear(x, w, b):
    return np.dot(x, w) + b

# derivative of the linear function - linear'(x) = x   // I did NOT use this function later  
def linear_derivative(x, w, b):                        # // because it worsened my code in terms of readibility  
    return x
```

```python
# DEFINITON OF AUTOENCODER PROPOGATIONS
# forward propagation of the autoencoder
def forward(X):
    u1 = linear(X, W1, b1)    # u1 = X * W1 + b1
    hidden1 = sigmoid(u1)    # hidden1 = sigmoid(u1)
    u2 = linear(hidden1, W2, b2)    #u2 = hidden1 * W2 + b2
    hidden2 = sigmoid(u2)    # hidden2 = sigmoid(u2)
    output = linear(hidden2, W3, b3)    # output = hidden2 * W3 + b3 
    return hidden1, hidden2, output    # returns hidden layers (hidden1, hidden2) and the output layer 

# backward propagation of the autoencoder to calculate the gradients
def backward(X, hidden1, hidden2, output, target):    # output is the output obtained from the forward function
    error = output - target    # error = forward(X) - target
    dW3 = np.dot(hidden2.T, error) # gradients of output layer
    db3 = np.sum(error, axis=0) # linear_derivative(error, W3, b3) = error
    
    error = np.dot(error, W3.T) * sigmoid_derivative(hidden2)
    dW2 = np.dot(hidden1.T, error) # gradients of second hidden layer
    db2 = np.sum(error, axis=0) # linear_derivative(error, W2, b2) = error 
    
    error = np.dot(error, W2.T) * sigmoid_derivative(hidden1)
    dW1 = np.dot(X.T, error) # gradients of first hidden layer
    db1 = np.sum(error, axis=0) # linear_derivative(error, W1, b1)
    return dW1, db1, dW2, db2, dW3, db3    # return all calculated gradients

# encoder function
def encoder(X):
    hidden1, hidden2, temp = forward(X)
    return hidden1, hidden2

# decoder function
def decoder(hidden1, hidden2):
    temp1, temp2, output = forward(X)
    return output
```

```python
# we can set number of epochs and learning rate here
num_epochs = 1000
learning_rate = 0.000001

X_train = X_training.values
loss = 0

# train the autoencoder with obtained gradients
for epoch in range(num_epochs):
    for i in range(len(X_train)):
        X = X_train[i].reshape((1, input_size))
        hidden1, hidden2, output = forward(X)
        dW1, db1, dW2, db2, dW3, db3 = backward(X, hidden1, hidden2, output, X)
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2
        W3 -= learning_rate * dW3
        b3 -= learning_rate * db3
        loss += np.sum((output - X) ** 2)
    if(epoch % (num_epochs / 10) == 0):    # print in every %10 progress
        print("Epoch: " , epoch , "Loss: ", loss / len(X_train))
    loss = 0
```

```python
# reduce the dimensionality of the data using the encoder function
hidden_1, hidden_2 = encoder(X_testing)

# reconstruct the data using decoder function
X_decoded = decoder(hidden_1, hidden_2)

# calculate the reconstruction error
reconstruction_error = np.sum((X_decoded - X_testing) ** 2) / len(X_testing)
error_sum = reconstruction_error.sum().sum()
print("Reconstruction Error: " , error_sum)
```

```python
# get hidden layers for the training data and concatenate them with labels
hidden1, hidden2, _ = forward(X_training)
reduced_training = np.concatenate((hidden1, hidden2, y_training.values.reshape(-1, 1)), axis=1)   

# same for the test data
hidden1, hidden2, _ = forward(X_testing)
reduced_testing = np.concatenate((hidden1, hidden2, y_testing.values.reshape(-1, 1)), axis=1)

# plot of the reduced data
plt.scatter(reduced_training[:, 0], reduced_training[:, 1], c=reduced_training[:, 2])
plt.show()

plt.scatter(reduced_testing[:, 0], reduced_testing[:, 1], c=reduced_testing[:, 2])
plt.show()
```
