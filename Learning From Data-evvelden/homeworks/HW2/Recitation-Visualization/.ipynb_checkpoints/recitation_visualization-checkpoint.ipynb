### Line Plots

Basically, these plots represent $y = f(x)$ in the x-y plane.

```python
x = np.linspace(0, 10, 100) # linspace divides the given interval into 100 equally spaced numbers:
                            # it goes like [0.        ,  0.1010101 ,  0.2020202 ,  0.3030303 ,  0.4040404 ,
                            # 0.50505051, ..., 9.5959596 ,  9.6969697 ,  9.7979798 ,  9.8989899 , 10.]       

fig = plt.figure("Cosine",figsize = (8,5)) # name the graph as a variable name
plt.plot(x, np.cos(x), color = "lightseagreen",label = "cos") # cosine of X
plt.plot(x, -np.cos(x), color="salmon", linestyle="dashdot",label = "--cos") # cosine of X
plt.plot(x,np.zeros(len(x)), color = "blue",linestyle="dashed",alpha=0.7,label = "zero") # alpha is the transparency
plt.legend(fontsize="x-large",frameon=True)
plt.show("Cosine")
plt.close("Cosine")
```

### Scatter Plots

In scatter plot, data points are represented by a marker (usually dots, but can also be triangle, cross and so on). Therefore, there is no continuity like line plot.

```python
from sklearn.datasets import load_iris
iris = load_iris()
feature_names = iris.feature_names
print(feature_names) # ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
data = iris.data
```

```python
iris_df = pd.DataFrame(data=data,columns=feature_names) # create a DataFrame from an array with given column names
iris_df["target"] = iris.target
iris_df.head()
```

In below plot, relationship between sepal and petal length can be viewed in the 2-D plane. Also, size of the shown data points represent petal width. A new legend is added to present the what petal width value corresponds to data point according to size in the plot. To do this, empty scatter is plotted.

```python
plt.figure(figsize=(12,8))
ax = plt.axes()
scatter = ax.scatter(iris_df['sepal length (cm)'].values,iris_df['petal length (cm)'].values,
                     s = iris_df['petal width (cm)'].values * 100, 
           c = iris_df.target,cmap="viridis", alpha=0.2)
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper left", title="Classes")
ax.add_artist(legend1)
# a empty scatter trick just for second legend
for area in [0.2, 1, 2]:
    ax.scatter([], [], c='k', alpha=0.3, s=area*100,
                label=str(area))
# new legend guides the observer about the meaning of the size of the data point
legend2 = ax.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Petal Width',loc = "lower right")
ax.add_artist(legend2)
ax.set_title("Sepal Length vs Petal Length")
ax.set_xlabel("sepal length (cm)")
ax.set_ylabel("petal length (cm)")
# OR
#  ax.set(xlabel='sepal length (cm)', ylabel='petal length (cm)',
#        title='Sepal Length vs Petal Length')
plt.show()
```

When object-oriented coding is used (axes) syntax differs as it can be seen when setting title, xlabel and ylabel. Be aware of that! Below is the function-based approach.

```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(iris_df['petal length (cm)'].values,iris_df['petal width (cm)'].values,s = iris_df['sepal length (cm)'].values * 100, 
           c = iris_df.target,cmap="viridis", alpha=0.2)
plt.legend(*scatter.legend_elements(),
                    loc="upper left", title="Classes")
plt.title("Petal Length vs Petal Width")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.show()
```

You can limit the plot into specified value interval if you want to make observer focus on a specific region for example.

```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(iris_df['petal length (cm)'].values,iris_df['petal width (cm)'].values,s = iris_df['sepal length (cm)'].values * 100, 
           c = iris_df.target,cmap="viridis", alpha=0.2)
plt.legend(*scatter.legend_elements(),
                    loc="upper left", title="Classes")
plt.xlim(0,2) # set limit for x
plt.ylim(0,1) # set limit for y
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal length (cm)")
plt.ylabel("petal width (cm)")
plt.xticks()
plt.show()
```

You may change the markers for each class to make them more seperable through the eye.

```python
classes = np.unique(iris_df.target)
markers = ['o','x','*']
colors = ["purple","black","yellow"]
plt.figure(figsize=(12,8))
for idx,c in enumerate(classes):
    iris_df_1class = iris_df[iris_df.target==c]
    scatter = plt.scatter(iris_df_1class['petal length (cm)'].values,iris_df_1class['petal width (cm)'].values,
                          s = iris_df_1class['sepal length (cm)'].values * 100, 
           c = colors[idx] , alpha=0.2,marker=markers[idx], label=c)
plt.legend(loc="upper left", title="Classes",fontsize="xx-large")
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal length (cm)")
plt.ylabel("petal width (cm)")
plt.xticks()
plt.show()
```

### Contour Plots

Sometimes we need to present three-dimensional data in two-dimensions. This can be done with contour plots.

```python
plt.style.use('seaborn-white')
```

Consider a function of $z = f(x,y)$. How to represent z with contour plot?

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

Create a meshgrid of two inputs. Meshgrid is two-dimensional grid of two inputs. It helps us to create paired inputs for coordinate matrices:

![resim.png](meshgrid.png)

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

### Histograms

Histogram is a plot type that presents the occurence/frequency of values. 

```python
random_data = np.random.randn(10000) # generate random data which comes from standard normal distribution (we expect bell curve)
```

```python
random_data.mean() # ~0
```

```python
random_data.std() # ~1
```

```python
plt.figure()
plt.hist(random_data,bins=20) 
plt.show() # normal distribution can be clearly seen. Bigger portion of the data lies around the mean (0)
```

```python
plt.figure()
plt.hist(random_data,bins=20,density=True) # if normed=True, occurence values transformed into occurence probability values 
#  (density = counts / (sum(counts) * np.diff(bins)))
plt.show() # normal distribution can be clearly seen. Bigger portion of the data lies around the mean (0)
```

#### Exercise: 
You are the lecturer that offers MAT101 class with 3 sessions a week. You want to analyze midterm grades visually for all sessions at once. How would you do?

-- 3 graphs in 1 plot with nice parameter settings!

```python
class1 = np.random.normal(50,10,100)
class2 = np.random.normal(75,10,100)
class3 = np.random.normal(85,7,100) # outstanding students
```

As you can see, all of the midterm grades are unique and occured only once in the class. Histogram creates bins as given number and keep these values into corresponding bin.

```python
np.unique(class1,
         return_counts=True)
```

```python
plt.figure(figsize=(12,8))
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=20)
plt.title("Midterm Grade Distrubition for 3 Classes",fontsize="x-large")
hist1 = plt.hist(class1, **kwargs,label="class1")
hist2 = plt.hist(class2, **kwargs,label="class2")
hist3 = plt.hist(class3, **kwargs,label="class3")
plt.xlabel("Midterm Grade",fontsize="x-large")
plt.ylabel("Occurence/Frequency",fontsize="x-large")
plt.legend(fontsize="x-large")
```

Here first element of the tuple represents how many data points are in that bin. Corresponding bin interval is written in the second element of the tuple. If value is greater than curent bin's value and less than the next bin's value, it belongs to that bin. For example, first bin's value is 26.34458557 for class1. This bin has 3 data points: 26.34458557, 28.0723988 , 28.1717683. Because these are greater than 26.34458557 (current bin value) and less than 28.72587817(next bin value).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

Below example is directly excerpted from the textbook:

```python


import pandas as pd
cities = pd.read_csv('california_cities.csv')

# Extract the data we're interested in
lat = cities['latd'].values
lon = cities['longd'].values
population = cities['population_total'].values
area = cities['area_total_km2'].values

```

```python
# 1. Draw the map background
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='i', 
            lat_0=37.5, lon_0=-119,
            width=1E6, height=1.2E6)
m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# 2. scatter city data, with color reflecting population
# and size reflecting area
m.scatter(lon, lat, latlon=True,
          c=np.log10(population), s=area,
          cmap='Reds', alpha=0.5)

# 3. create colorbar and legend
plt.colorbar(label=r'$\log_{10}({\rm population})$')
plt.clim(3, 7)

# make legend with dummy points
for a in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.5, s=a,
                label=str(a) + ' km$^2$')
plt.legend(scatterpoints=1, frameon=False,
           labelspacing=1, loc='lower left');
```

### Read file

```python
df = pd.read_csv("automobile.csv")
print(df.columns)
```

```python
df.drop(["symboling","normalized-losses"],axis=1) # drop unnecessary columns
```

```python
df.replace('?',np.nan,inplace=True) # replace ? with NaN for making the unknown cells standard empty.
df.dropna(inplace=True,axis=0,how="any") # now we can directly drop rows containing NaN (previously ""?"")
```

```python
df.head()
```

Matplotlib treats strings as it is. Unfortunately, horsepower colum is set as object type of object not a numerical type of object due to unknown("?") string values. That is why below plot is not visually appealing and representative. Labels on ticks are collided with each other such that graph is not readable :( Let's fix it in upcoming cells

```python
plt.figure()
plt.scatter(df["horsepower"],df["price"])
plt.show()
```

Matplotlib treats strings as it is. Unfortunately, horsepower colum is set as object type of object not a numerical type of object due to unknown("?") string values. Let's fix it to create plots:

```python
df["horsepower"] = df["horsepower"].astype(int) # change to integer type
df["price"] = df["price"].astype(int)  # change to integer type
```

```python
plt.figure()
plt.title("Effect of Horsepower on Price")

plt.scatter(df["horsepower"],df["price"])
plt.xlabel("Horsepower")
plt.ylabel("Price")
plt.show()
```

```python
cross_tab = pd.crosstab(
    df["fuel-type"].values,df["body-style"].values, margins=True)
```

```python
sns.heatmap(cross_tab)
```

Plot a bar graph that presents average price ("price" column) for each brand of car ("make" column)

```python
groupby_make = df.groupby("make")["price"].mean()
groupby_make
```
