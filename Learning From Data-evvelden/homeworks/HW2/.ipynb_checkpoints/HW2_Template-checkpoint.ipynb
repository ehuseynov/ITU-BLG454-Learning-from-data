## **2.** Compute the linear discriminant function gi(x) for each class regarding the following cases:**(35 points)**
<br>
<br>
$g_i(x) = -1/2 * (x - \mu_i)^t * Σ_i^{-1} * (x - \mu_i) - 1/2 * ln|Σ_i| + ln(P(c_i))$
<br>
<br> 
since they are perfectly balanced, I can erase "+ $ln(P(c_i))$" ($P(c_1) = P(c_2)$) 
<br>
$g_i(x) = -1/2 * (x - \mu_i)^t * Σ_i^{-1} * (x - \mu_i) - 1/2 * ln(Σ_i)$
<br>

### Case 1 (25 points)**: Equal covariances across the classes.  ( $Σ_1$ = $Σ_2$)
<br>
$Σ_i$ = Σ
<br>    
~~1/2 * ln(Σ_i)~~ because all $Σ_i$'s are even 
<br>
$g_i(x) = -1/2 * (x - \mu_i)^t * Σ_i^{-1} * (x - \mu_i)$
<br>

```python
def discriminant_i_func_EVEN(x, mean_i, cov_i):
    # temp = x - mu_i
    part_temp = np.subtract(x,mean_i)
    part_a = np.transpose(part_temp)
    part_b = np.linalg.inv(cov_i)
    part_c = part_temp
    result = -1/2 * part_a @ part_b @ part_c
    return result

discriminant_i_func_EVEN(test1_data[0] , np.mean(df1_class1), np.cov(df1_class1.T) )
```

### **Case 2 (20 points)**: General case where the discriminant function $g_i$(x) is quadratic
<br>
General Formula: 
<br>
$g_i(x) = -1/2 * (x - \mu_i)^t * Σ_i^{-1} * (x - \mu_i) - 1/2 * ln|Σ_i| $
<br> 
<br>


```python
def discriminant_i_func_GENERAL(x, mean_i, cov_i):
    # temp = x - mu_i
    part_temp = np.subtract(x,mean_i)
    part_a = np.transpose(part_temp)
    part_b = np.linalg.inv(cov_i)
    part_c = part_temp
    part_d = np.linalg.det(cov_i)
    result = -1/2 * part_a @ part_b @ part_c * -1/2 * math.log(part_d)
    return result

discriminant_i_func_GENERAL(test1_data[0] , np.mean(df1_class1), np.cov(df1_class1.T) )
```

```python
def trainBayes(test_x ,trainingSamples, trainingLabels):
    #This function seperates classes from trainingSamples
    #then depending on their covariance situatin makes relaterd function call
    #finally returns estimation of the class
    c1 = trainingSamples[trainingLabels==0]
    c2 = trainingSamples[trainingLabels==1]
    cov_1 =np.cov(c1.T)
    cov_2 = np.cov(c2.T)
    #print("Covariance matrix of class 1:\n",cov_1)
    #print("Covariance matrix of class 2:\n",cov_2)
    
    mean_1 = np.mean(c1, axis=0)
    mean_2 = np.mean(c2, axis=0)
    
    #print("Mean vector of class 1: ",mean_1)
    #print("Mean vector of class 2: ", mean_2)
    
    if(np.array_equal(cov_1,cov_2)):
        g1_x = discriminant_i_func_EVEN(test_x , np.mean(mean_1), cov_1 )
        g2_x = discriminant_i_func_EVEN(test_x , np.mean(mean_2), cov_2 )
        if(g1_x > g2_x):
            return 1
        else: 
            return 2
    else:
        g1_x = discriminant_i_func_GENERAL(test_x, np.mean(mean_1), cov_1 )
        g2_x = discriminant_i_func_GENERAL(test_x , np.mean(mean_2), cov_2 )
        if(g1_x > g2_x):
            return 1
        else: 
            return 2
```

```python
bayes = trainBayes(test1_data[0],df1_data, df1_labels)
print("It belongs to class: " , bayes)    # debug line
```

## Part C: Implementation of PCA (25 Points)

```python
#Implement PART C:

def PCA(X, k=None, varRetained = 0.95 , show = False ) :
    X = X - X.mean()    # center the data
    cov_matrix = np.cov(X.T)   # covariance matrix of given dataset
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)    # eigenvalues and eigenvectors of given dataset
    idx = eigenvalues.argsort()[::-1]    # sorting eigenvalues in order to choose biggest one
    eigenvalues = eigenvalues[idx]    # biggest eigenvalue
    eigenvectors = eigenvectors[:,idx]    # corresponding eigenvector
    cum_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)   # cumulative sum of the variances
    
    if k is None:
        k = np.where(cum_var >= varRetained)[0][0] + 1
        
    projection_matrix = eigenvectors[:, :k]    # top k principal components
    
    # to get new values (project into a new dataset), we multiply given dataset with projection matrix
    reduced_feature_set = X.dot(projection_matrix)   

    # visualization of variance vs. number of features if show == true
    if show:
        plt.plot(range(1, len(eigenvalues) + 1), cum_var)
        plt.title("PCA Variance vs # Features")
        plt.xlabel('Number of Features')
        plt.ylabel('Percentage Variance retained')
        plt.show()

    return projection_matrix, reduced_feature_set
```

```python
# IRIS DATASET
from sklearn.datasets import load_iris
iris = load_iris()

iris_data = iris.data   # iris data in numpy array format
iris_targets = iris.target    # iris labels in numpy format

# to learn more information about the dataset
print(type(iris_data))
print(iris_data.shape)
print(iris_targets.shape)

print("IRIS DATASET\n")
# BEFORE PCA
plt.scatter(iris_data[:, 0], iris_data[:, 1], c=iris_targets)
plt.title("Iris Dataset Before PCA")
plt.show()

k = 2 # to easily set the k value
procection_matrix , iris_data_after_pca = PCA(iris_data, k, 1, 1)    # PCA

# AFTER PCA
plt.scatter(iris_data_after_pca[:, 0], iris_data_after_pca[:, 1], c=iris_targets)
plt.title("Iris Dataset After PCA")
plt.show()
```
