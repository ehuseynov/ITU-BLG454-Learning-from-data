### Read file

```python
df = pd.read_csv("automobile.csv")
print(df.columns)
```

```python
df.drop(["symboling","normalized-losses"],axis=1) # drop unnecessary columns
```

```python
df.replace('?',np.nan,inplace=True) # replace ? with NaN for making the unknown cells standard empty.
df.dropna(inplace=True,axis=0,how="any") # now we can directly drop rows containing NaN (previously ""?"")
```

```python
df.head()
```

Matplotlib treats strings as it is. Unfortunately, horsepower colum is set as object type of object not a numerical type of object due to unknown("?") string values. That is why below plot is not visually appealing and representative. Labels on ticks are collided with each other such that graph is not readable :( Let's fix it in upcoming cells

```python
plt.figure()
plt.scatter(df["horsepower"],df["price"])
plt.show()
```

Matplotlib treats strings as it is. Unfortunately, horsepower colum is set as object type of object not a numerical type of object due to unknown("?") string values. Let's fix it to create plots:

```python
df["horsepower"] = df["horsepower"].astype(int) # change to integer type
df["price"] = df["price"].astype(int)  # change to integer type
```
