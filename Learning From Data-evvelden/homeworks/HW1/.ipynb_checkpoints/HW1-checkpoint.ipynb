# Part 1: Simple Linear Regression  (30 points)

### Background

Remember from the course material: In Linear Regression, we aim to find the line that fits the given data as well as possible. The line formula is:

### <center> $ \hat{y} = f(x) = \beta_0 + \beta_1 x $ <center> 

And using linear regression, the goal is to find the $\beta_0$ and $\beta_1$ such that the the Mean Squared Error (MSE) is minimized:
    
### <center> $ MSE \Rightarrow \mathcal{L} = \dfrac{1}{N} \sum_{i=1}^{N} (y_i - f(x_i ))^2 $ <center> 
    
As the MSE error gets smaller, the line starts to represent the data better and better.
    
Consider $\beta = (\beta_0 + \beta_1)$. In order to find the $\beta$ that minimizes the MSE loss, we use **Gradient Descent**. Iteratively $\beta$ is updated in the opposite direction of the gradient, while the size of the update is controlled with the learning rate $\eta$:
    
### <center> $ \beta \leftarrow \beta-\eta \frac{d \mathcal{L}}{d \beta} $ <center>

```python
def load_data():
    df = pd.read_csv("data1.csv")
    x = df['x'].values
    y = df['y'].values
    return x, y


def gradient_descent(x, y, learning_rate = 0.1, max_its = 500):
    #TO DO: Initialize the beta values
    beta0 = 1
    beta1 = 1
    
    # you can change the num_of_prints to see more steps of the calculation process
    num_of_prints = 10
    step_to_print = max_its / num_of_prints
    step = 0
    
    # start gradient descent loop
    for k in range(0,max_its):
        # TO DO: calculate derivative (grad)
        X ,Y = load_data();
        Y_hat = (X * beta0) + beta1
        error = Y - Y_hat             # my error function
        mse = np.square(error).mean() # mean squared error function
        
        gradient_beta0 = - (1 / len(X)) * np.dot(X.T, error)
        gradient_beta1= - (1 / len(X)) * np.sum(error)   
        
        # TO DO: take gradient descent step
        beta0 = beta0 - learning_rate * gradient_beta0
        beta1 = beta1 - learning_rate * gradient_beta1
        
        # TO DO: calculate and print the loss value
        ...
        step = step + 1
        
        if(step_to_print < step):
            print(np.sum(mse))
            step = 0

    return beta0, beta1
```

```python
x, y = load_data()
beta0, beta1 = gradient_descent(x, y)

# TO DO: Plot the data AND the line!

plt.plot(x, y, 'o')
plt.plot(x, beta0 * x + beta1 )
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import KFold
def load_data():
    data= fetch_california_housing()
    print("Features of Boston Housing Prices dataset: ", data.get("feature_names"))
    print("Shape of the X: ", data.data.shape)
    print("Shape of the Y: ", data.target.shape)
    return data.data, data.target



def multiple_gradient_descent_KFold(k_parameter = 5,learning_rate = 0.1, max_its = 10):
    X, Y = load_data()
    n = len(Y)
    
    num_of_prints = 10
    print_step = max_its / num_of_prints
    print("It will print every",print_step, "step so be patient")
    
    #in this for loop I split total dataset into 5 and in each iteration I took the next test set
    mses = [] 
    for split in range(0,5):
        print("")
        print("For fold=",split,"we have: ")
        
        # made split operations by hand because I don't know any better
        X_test = X[int(split * n / 5): int((split + 1) * n / 5),:]
        Y_test = Y[int(split * n / 5): int((split + 1) * n / 5)]
        
        X_train_1 = X[:int(split * n / 5),:]
        X_train_2 = X[int((split + 1) * n / 5):,:]
        X_train = np.row_stack((X_train_1,X_train_2))
        
        Y_train_1 = Y[:int(split * n / 5)]
        Y_train_2 = Y[int((split + 1) * n / 5):]
        Y_train = np.concatenate((Y_train_1,Y_train_2))
    
        '''
        the structure I am going for:
        x0(i).beta0 + x1(i).beta1 + ...... x7(i).beta7 + bias = y(i)
        but there is no need to take the bias as a seperate parameter
        I can simple write:
        x0(i).beta0 + x1(i).beta1 + ...... x7(i).beta7 + x8(i).beta8= y(i)
        beta8 = bias and x8 = 1 always
        '''
        beta = np.ones([8,1])
        bias = np.ones(1)
            
        X_train = np.concatenate([X_train , np.ones((X_train.shape[0],1))], axis = 1)
        beta = np.concatenate([beta,bias.reshape(1,-1)],axis = 0)
        # now I have to add bias term to test set as well 
        X_test = np.concatenate([X_test , np.ones((X_test.shape[0],1))], axis = 1)
        
        step = print_step
        
        
        for i in range(0,max_its):
            Y_hat_train = X_train @ beta 
            error_train = Y_train - Y_hat_train            
            mse_train = np.square(error_train).mean()

            gradient_beta = - (1 / len(X_train)) * np.sum(X_train.T @ error_train)
            beta = beta - (learning_rate * gradient_beta)
            
            Y_hat_test = X_test @ beta 
            error_test = Y_test - Y_hat_test            
            mse_test = np.square(error_test).mean()
            
            step = step - 1
            if(step < 1):
                print("mse:", mse_test)
                step = print_step
            if(i == max_its-1):
                mses.append(mse_test)
           
        
        
    mses = np.array(mses)
    print("")
    print("Mean of mse's:", np.mean(mses))   
    print("Variance:" , np.var(mses))
    '''
    For multi linear regression I made what I did previously on simple linear regression
    but with 8 features and making use of matrix. 
    For some reason with bigger learning rates mse goes larger and larger. That's why I 
    pick the learning rate this low.
    '''
        
    return
```
